# Help Functions

```python
def lcm(a,b): 
    return abs(a * b) / fractions.gcd(a,b) if a and b else 0 

def get_multi(x):
    S = str(x).split(".")
    if (len(S) == 1):
        return 1
    L = 10**len(S[1])
    div = alldividers(L)
    for i in div:
        if(i*x == float(int(i*x))):
           return i
                      
def alldividers(zahl):
    multi = 10.0
    try:
        temp = str(zahl).split(".")
        multi = 10.0**len(temp[1])
    except:
        pass        
    result = []
    for i in range (1, zahl + 1):
        if zahl * multi / i % multi == 0:
            result.append(i)
    return result


def create_structure(path, a, c, a11 ,a12 ,a21 ,a22, l1, l2, bl, size = 0.0, zoffset1 = 0.0, zoffset2 = 0.0):
    f = open(path, 'w')
    random.seed()
    L1 = get_multi(l1)
    L2 = get_multi(l2)
    L = lcm(L1,L2)
    layers = bl*(l1+l2)
    if (float(int(layers)) == layers):
        layers = int(layers)
    else:
        layers = int(layers)+1
    f.write('a ' + str(a*L*10**(-10)) + '\n')
    f.write('b ' + str(a*10**(-10)) + '\n')
    f.write('c ' + str(c*layers*10**(-10)) + '\n')
    atom = 1
    l = l1
    offset = 0
    zoffset = zoffset1
    for i in range(layers):
        for j in range(L):
            #check that the right atom is printed
            if (i*L+j - offset >= l*L):
                if (atom == 1):
                    l = l2
                    atom = 2
                    zoffset += zoffset2
                else:
                    l = l1
                    atom = 1
                offset = i*L+j
                zoffset += zoffset1
            if (atom == 1):
                f.write('element ' + a11 + '\n')
            else:
                f.write('element ' + a21 + '\n')
            f.write('x ' + str((j+random.randrange(-1,1)*size)/L) + '\n')
            f.write('y ' + str(0.0+random.randrange(-1,1)*size) + '\n')
            f.write('z ' + str(float(i+1+zoffset+random.randrange(-1,1)*size)/layers) + '\n')
            if (atom == 1):
                f.write('element ' + a12 + '\n')
            else:
                f.write('element ' + a22 + '\n')
            f.write('x ' + str((j+0.5+random.randrange(-1,1)*size)/L) + '\n')
            f.write('y ' + str(0.5+random.randrange(-1,1)*size) + '\n')
            f.write('z ' + str((i+0.5+zoffset+random.randrange(-1,1)*size)/layers) + '\n')
            f.write('element O' + '\n')
            f.write('x ' + str((j+0.5+random.randrange(-1,1)*size)/L) + '\n')
            f.write('y ' + str(0.5+random.randrange(-1,1)*size) + '\n')
            f.write('z ' + str(float(i+1+zoffset+random.randrange(-1,1)*size)/layers) + '\n')
            f.write('element O' + '\n')
            f.write('x ' + str((j+0.5+random.randrange(-1,1)*size)/L) + '\n')
            f.write('y ' + str(0.0+random.randrange(-1,1)*size) + '\n')
            f.write('z ' + str((i+0.5+zoffset+random.randrange(-1,1)*size)/layers) + '\n')
            f.write('element O' + '\n')
            f.write('x ' + str(float(j+random.randrange(-1,1)*size)/L) + '\n')
            f.write('y ' + str(0.5+random.randrange(-1,1)*size) + '\n')
            f.write('z ' + str((i+0.5+zoffset+random.randrange(-1,1)*size)/layers) + '\n')
    f.close()
    print "l", L, "layers", layers
    
def showup(l1, l2):
    L1 = get_multi(l1)
    L2 = get_multi(l2)
    L = lcm(L1,L2)
    bl = get_multi((l1+l2))
    layers = float(bl)*(l1+l2)
    
    if (float(int(layers)) == layers):
        layers = int(layers)
    else:
        print "error bilayers wrong"    
    
    atom = 1
    l = l1
    offset = 0
    result = ''
    for i in range(layers):
        for j in range(L):
            #check that the right atom is printed
            if (i*L+j - offset >= l*L):
                if (atom == 1):
                    l = l2
                    atom = 2
                else:
                    l = l1
                    atom = 1
                offset = i*L+j
            if (atom == 1):
                result += 'B'
            else:
                result += 'S'
        result += '\n'
    each = result.split('\n')
    temp = []
    for i in range(len(each)-1, -1 ,-1):
        temp.append(each[i])
    for line in temp:
        print line
```

# defining Experiment

```python
l_min = 0.8
l_max = 1.05
l_step = 0.0002 #5000 points per order of l

wavelength = 1.5409E-10
direct = 2.0E8
background = 1

q = [0.0,0.0,0.0]
steps_l = (l_max-l_min)/l_step

STO = xpy.CrystalStructureCheck(path = xpy.scriptpath+"Perovskites/STO.str")
substrate = xpy.ThickFilmCheck(STO)
```

# Compare 50 layer thinfilm with 50 single layer films

```python
BTO = xpy.CrystalStructureCheck(path = xpy.scriptpath+"Perovskites/PTOs.str")
filmBTO = xpy.ThinFilm(BTO,1)
BTOfilm = xpy.ThinFilm(BTO,50)
filmBTO.delta = 0.
SuperBTOfilm = xpy.SuperLattice(filmBTO,filmBTO,25)

sBTO001 = []
BTO001 = []
q[0] = 0.0#2*m.pi/STO.lattice_parameters[0]
q[1] = 0.0
for i in range(int(steps_l)):
    l=l_min+i*l_step
    q[2] = l*2*m.pi/STO.lattice_parameters[2]
    absq = sum([x**2 for x in q])
    sinomega = 4.0 * m.pi / absq / wavelength
    sub = substrate.get_reflection(q, sinomega)
    lin1 = abs(
              SuperBTOfilm.get_reflection(q, sinomega)  
              + sub * SuperBTOfilm.get_phase_shift(q, sinomega)
             )
    lin1 *= lin1
    lin1 *= direct
    lin1 += background
    sBTO001.append([l, lin1])
    
    lin1 = abs(
              BTOfilm.get_reflection(q, sinomega)  
              + sub * BTOfilm.get_phase_shift(q, sinomega)
             )
    lin1 *= lin1
    lin1 *= direct
    lin1 += background
    BTO001.append([l, lin1])
```

```python
fig = plt.figure(figsize=(16,10), dpi=100)

axes = fig.add_axes([0., 0., 1, 1]) # left, bottom, width, height (range 0 to 1)


axes.plot([x[0] for x in sBTO001], [y[1] for y in sBTO001], color="blue", lw=1, ls='*', marker='+')

axes.plot([x[0] for x in BTO001], [y[1] for y in BTO001], color="green", lw=1, ls='*', marker='+')
axes.plot([x[0] for x in BTO001], [abs(float(a[1]) - float(b[1])) for a, b in zip(sBTO001, BTO001)], color="red", lw=1, ls='*', marker='+')
axes.plot([x[0] for x in BTO001], [(abs(float(a[1]) - float(b[1])))/float(a[1]) for a, b in zip(sBTO001, BTO001)], color="blue", lw=1, ls='*', marker='+')

axes.set_xlabel('l')
axes.set_ylabel('int[A.U.]')
axes.set_title('title')
axes.set_yscale("log")
```
