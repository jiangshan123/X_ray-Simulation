```python
path=r'/Users/Rui/Dropbox/Research/STONY BROOK/Matt Dawber/Python Simulation/PTO BTO Project/Growth/J0420161'
data = []
f = open(path, 'r', encoding='iso-8859-15')
line=f.readline()
while (line):
    words = line.split('\t')
    if words:
        if (len(words) > 1):
            data.append(words)
    line=f.readline()
f.close()   
```

```python
new1 = []
for val in data[0]:
    if val != '':
        new1.append(val)
data[0] = new1

new2 = []
for val in data[1]:
    if val != ' ':
        new2.append(val)
data[1] = new2
data[1].insert(0, '')

dic = {}
for key, val in zip(data[0],data[1]):
        if val == '': 
            val = []
        dic[key] = val

data.pop(0)   
data.pop(0)
    
```

```python
for point in data:
    dic['Time (min)'].append(point[0])
    dic['Temperature (째C)'].append(point[1])
    dic['Pressure (mTorr)'].append(point[2])
    dic['Position'].append(point[3])
    dic['Flow O2'].append(point[4])
    dic['Flow Ar'].append(point[5])
    dic['Active Gun'].append(point[6])
    dic['Forward Power (W)'].append(point[7])
    dic['Reflected Power (W)'].append(point[8])
    dic['Sequence'].append(point[9])
    dic['Repetition'].append(point[10])
    dic['Voltage'].append(point[11])
```

```python
for key in dic:
    print(key)
```

```python
class Gun(object):
    
    def __init__(self, Rate, pressure, Flow_AR, Flow_O, T_V, material, power):
        """initiakizes the class data
        """
        self.rate = Rate
        self.pressure = pressure
        self.material = material
        self.power = power
        self.temperature = T_V
        self.flow_ar = Flow_AR
        self.flow_o = Flow_O
        
    def __str__(self):
        """Returns the data in a structured form.
        """
        
        stri = ("Material:" + self.material + "\n" +
                "Growth Rate (min/Layer):" + self.rate + "\n" +
                "Temperature (째C) / Voltage (V):" + self.temperature + "\n" +
                "Pressure (mTorr):" + self.pressure + "\n" +
                "Power:" + self.power + "\n"
                "Ar Flow rate (ccm):" + self.flow_ar + "\n"
                "O2 Flow rate (ccm):" + self.flow_o + "\n"
                )
        return stri
```
